```python
import numpy as np
import matplotlib.pyplot as plt

def target_function(x):
    return ((1 + 0.6 * np.sin(2 * np.pi * x / 0.7)) + 0.3 * np.sin(2 * np.pi * x)) / 2

x = np.arange(0.1, 1.0, 1/220)

y_true = target_function(x)


plt.plot(x, y_true)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from random import random

# Define activation functions and their derivatives
def activation_function(x):
    return 1 / (1 + np.exp(-x))

def activation_derivative(x):
    return x * (1 - x)

# Initialize parameters
learning_rate = 0.1
epochs = 10000

# Generate input data
x = np.arange(0.1, 1.0, 1/220)
y_true = target_function(x)

# Initialize weights and biases for all layers
w11_1 = random()
b1_1 = random()
w11_2 = random()

w21_1 = random()
b2_1 = random()
w21_2 = random()

w31_1 = random()
b3_1 = random()
w31_2 = random()

w41_1 = random()
b4_1 = random()
w41_2 = random()



# Train the neural network using backpropagation
for epoch in range(epochs):
    # Forward pass
    v11_1 = x * w11_1 + b1_1
    y1_1 = activation_function(v11_1)
    v11_2 = y1_1 * w11_2

    v21_1 = x * w21_1 + b2_1
    y2_1 = activation_function(v21_1)
    v21_2 = y2_1 * w21_2

    v31_1 = x * w31_1 + b3_1
    y3_1 = activation_function(v31_1)
    v31_2 = y3_1 * w31_2

    v41_1 = x * w41_1 + b4_1
    y4_1 = activation_function(v41_1)
    v41_2 = y4_1 * w41_2

    y2 = v11_2 + v21_2 + v31_2 + v41_2  # Linear activation function in the output

    # Calculate error
    error = y_true - y2

    # Backward pass
    delta2 = error
    delta1_2 = delta2 * w11_2 * activation_derivative(y1_1)
    delta1_1 = delta2 * w21_2 * activation_derivative(y2_1)
    delta1_3 = delta2 * w31_2 * activation_derivative(y3_1)
    delta1_4 = delta2 * w41_2 * activation_derivative(y4_1)

    # Update weights and biases
    w11_2 += learning_rate * y1_1 * delta2
    w21_2 += learning_rate * y2_1 * delta2
    w31_2 += learning_rate * y3_1 * delta2
    w41_2 += learning_rate * y4_1 * delta2

    w11_1 += learning_rate * x * delta1_2
    b1_1 += learning_rate * delta1_2
    w21_1 += learning_rate * x * delta1_1
    b2_1 += learning_rate * delta1_1
    w31_1 += learning_rate * x * delta1_3
    b3_1 += learning_rate * delta1_3
    w41_1 += learning_rate * x * delta1_4
    b4_1 += learning_rate * delta1_4


# Plot the results
plt.plot(x, y_true, label='True Function')
plt.plot(x, y2, label='Predicted Function')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
